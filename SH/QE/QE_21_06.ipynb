```python
def smallest_positive_integer(arr):
    n = len(arr)
    
    # Iterate over the array and move all non-positive integers to the end
    j = 0
    for i in range(n):
        if arr[i] <= 0:
            arr[i], arr[j] = arr[j], arr[i]
            j += 1
    
    # Consider only the positive integers
    arr = arr[j:]
    n = len(arr)
    # print(arr) # [7, 4, 1, 2]
    
    # If all integers are non-positive, return 1
    if n == 0:
        return 1
    
    # Mark the presence of an integer by making the corresponding index negative
    for i in range(n):
        if abs(arr[i]) <= n: # e.g. if 7, the index to negative value
            arr[abs(arr[i]) - 1] = -abs(arr[abs(arr[i]) - 1])
    # print(arr) # [-7, -4, 1, -2]
    # Find the smallest positive integer that is not marked
    for i in range(n):
        if arr[i] > 0:
            return i + 1
    
    # If all integers from 1 to n are present, return n + 1
    return n + 1

# Test cases
array1 = [4, 7, -1, 1, 2]
array2 = [100, 101, 102]
array3 = [3, 2, 1, 0, -1]

print(smallest_positive_integer(array1))  # Output: 3
print(smallest_positive_integer(array2))  # Output: 1
print(smallest_positive_integer(array3))  # Output: 4

```

```python
class LinkedNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def swap(node1, node2):
    tmp = node2.data
    node2.data = node1.data
    node1.data = tmp


def sortingLL(head: LinkedNode) -> LinkedNode:
    swapped = True
 
    while swapped:
        swapped = False
        current = head
 
        while current.next:
            if current.data > current.next.data:
                swap(current, current.next)
                swapped = True
            current = current.next
    
```
