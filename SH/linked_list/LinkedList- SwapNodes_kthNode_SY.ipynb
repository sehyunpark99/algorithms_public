```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def swapNodes(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        first = second = head
        for _ in range(k - 1):
            first = first.next

        tail = first
        #"first"ê° ì´ë¯¸ k-1ë§í¼ ìì¼ë¡ ì´ëíê³ , "tail"ë "first"ì ê°ì ìì¹ìì ììíê¸° ëë¬¸ì "tail"ì´ ë¦¬ì¤í¸ì ëì ëë¬íë©´ "second"ë length - kë§í¼ ì´ëí ìì¹ì ìê²ë¨
        while tail.next:
            second = second.next
            tail = tail.next 

        first.val, second.val = second.val, first.val
        return head
```
