### 445. Add Two Numbers



```python
class Solution:
    # To reverse the order to linked list
    def reverse(self, node: ListNode) -> ListNode:
        # Initialization: prev-nothing, curr-first node of the list
        prev, curr = None, node
        # curr will be the one moving
        while curr:
            # one step further
            next, curr.next = curr.next, prev
            # current node to next node and previous node to current node
            prev = curr
            curr = next
        # curr will be None after all (like moving curr to prev)
        return prev
    
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        # 1. Reversing the ListNode
        l1 = self.reverse(l1)
        l2 = self.reverse(l2)

        # 2. Addition of two numbers (02. Add Two Numbers)
        dummy = ListNode(0)
        tail = dummy
        carry = 0
				# All the conditions should be specified
        while l1 or l2 or carry: 
            num1 = l1.val if l1 else 0
            num2 = l2.val if l2 else 0
            # Result of the summation
            num = num1+num2+carry
            # Will store only first digits
            tail.next = ListNode(num%10)
            # Set the carry -> either 0 or 1
            carry = num//10
            tail = tail.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        dummy = self.reverse(dummy.next)
        return dummy
```

## Addition

```python
# Definition for singly-linked list.
def helper(self, l1: ListNode, l2: ListNode) -> ListNode:
        dummyHead = ListNode(0)
        tail = dummyHead
        carry = 0

        while l1 or l2 or carry:
            digit1 = l1.val if l1 else 0
            digit2 = l2.val if l2 else 0

            total = digit1 + digit2 + carry
            digit = total % 10
            carry = total // 10

            newNode = ListNode(digit)
            tail.next = newNode
            tail = tail.next

            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None

        result = dummyHead.next
        return result
```

## Reverse

```python
def reverseList(self, head: ListNode) -> ListNode:
        prev = None
        curr = head

        while curr:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
        
        return prev
```
