### 24. Swap Nodes in Pairs
https://leetcode.com/problems/swap-nodes-in-pairs/

1. head  :       (head.next = next.next)
2. next  :    (next.next = head)
3. prev  :      (prev.next = next)

 head  prev   swap  node  

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def swapPairs(self, head):
        root = prev = ListNode(0)
        prev.next = head #1
        while head and head.next:
            next = head.next #2
            head.next = next.next #1-3
            next.next = head #2-1

            prev.next = next #2

            head = head.next 
            prev = prev.next.next

        return root.next
```

```python
# Example
root = head = ListNode(0)
for i in [1,2,3,4]:
    head.next = ListNode(i)
#    print(head.next.val)
    head = head.next
newroot = root.next
```

```python
sol = Solution()
answer = sol.swapPairs(newroot)
```
