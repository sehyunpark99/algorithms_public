### 24. Swap Nodes in Pairs
https://leetcode.com/problems/swap-nodes-in-pairs/

1. head 기준 : 뒤 노드랑 새로 연결해 주는 역할 (head.next = next.next)
2. next 기준 : 자리를 바꿔주는 역할 (next.nextㅡ = head)
3. prev 기준 : 전 노드랑 연결해 주는 역할 (prev.next = next)

마지막으로 head 와 prev를 새로 지정해서 swap되지 않은 node들 기준으로 반복

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def swapPairs(self, head):
        root = prev = ListNode(0)
        prev.next = head #1
        while head and head.next:
            next = head.next #2
            head.next = next.next #1-3
            next.next = head #2-1

            prev.next = next #2

            head = head.next 
            prev = prev.next.next

        return root.next
```

```python
# Example
root = head = ListNode(0)
for i in [1,2,3,4]:
    head.next = ListNode(i)
#    print(head.next.val)
    head = head.next
newroot = root.next
```

```python
sol = Solution()
answer = sol.swapPairs(newroot)
```

```python
for _ in range(4):
    print(answer.val)
    answer = answer.next
```
