```python
#case1) 내장 모듈 쓰는 경우
from collections import Counter

def checkInclusion(self, s1: str, s2: str) -> bool:
        window = len(s1)
        s1_c = Counter(s1) #s1 철자별 count 담은 딕셔너리 반환

        
        for i in range(len(s2)-window+1):
            s2_c = Counter(s2[i:i+window]) 
            if s2_c == s1_c:
                return True
            
        return False
```

```python
#case2) dictionary 구현 
from collections import defaultdict

def checkInclusion(self, s1, s2):
        n = len(s1)
        cnt_dict = defaultdict(int)
        check_dict = defaultdict(int)
        ans = False
        
        for i in range(n):
            cnt_dict[s1[i]] += 1 #Counter 구현
        
        for j in range(len(s2)-n+1):
            for k in range(j,j+n):
                check_dict[s2[k]] += 1 #Counter 구현
            
            if cnt_dict.items() == check_dict.items():
                ans = True
                break
            
            check_dict = defaultdict(int) #check_dict 초기화 (=다시 새로 확인해야하므로)
            
        return ans
```
