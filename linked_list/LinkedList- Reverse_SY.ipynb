```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        curNode = head
        next = None
        while curNode:
            next = curNode.next 
            curNode.next = prev
            prev = curNode
            curNode = next
        return prev
```
